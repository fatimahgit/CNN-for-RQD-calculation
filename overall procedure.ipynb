```python
import glob
import os
from PIL import Image
import cv2
import numpy as np
from image_processing_functions import *
from RGD_calculation import *
from CNN_model import *
import torch 

model_path = "trained_model.pt"
class_to_idx_path = 'class_to_index.pth'

input_path = '../sandstone///*.jpg'
output_path = '../sandstone///output_report.csv' #


step = 100
device = torch.device('cpu')
model, class_to_idx = load_model(model_path,class_to_idx_path)

# creat lists for the results to be reported in the output file 
rows_names_total =[] 
depths_from_total = []   
depths_to_total = []  
RQDs_total = []

```

```python

import time

since = time.time()
for file in glob.glob(input_path):
       
    image = cv2.imread(file)
    image_name, ext = os.path.splitext(os.path.basename(file))
    print('image {} in process'.format(image_name))
    img= cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    extracted, script_vert, script_h, script_L_D, r1, r2 = main_extract(img) 
    depths_list = get_depth(script_vert, script_L_D)
    pixels_eq_10mm =get_scales(script_h)    
    
    RQDs, depths_from, depths_to, no_rows, rec= final(extracted, depths_list, step, pixels_eq_10mm ,model, class_to_idx,
                                        img, r1, r2) 

    rows_names = [image_name+'_row_'+str(i+1) for i in range (no_rows)]

    rows_names_total.extend(rows_names)
    depths_from_total.extend(depths_from)
    depths_to_total.extend(depths_to)
    RQDs_total.extend(RQDs)
print('RQDs', RQDs)

#rec = Image.fromarray(rec)
#rec.save(os.path.join(output_path,image_name+"4.jpeg"), rec.format)
output_report (output_path, rows_names_total, depths_from_total,depths_to_total, RQDs_total)    #after the for loop to includes all the images
print(time.time() - since)
print('done')
```
